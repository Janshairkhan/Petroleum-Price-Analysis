By plotting these data on a single graph, we can easily visualize the relationship between the two variables(Date & Petrol price) and identify any trends or patterns."

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a list to store the labels that have already been added
label_added = []

# Convert date column to datetime format
dframe_pkr['Date'] = pd.to_datetime(dframe_pkr['Date'])

# Plot line graph
sns.set(font_scale=2)
sns.set(rc={'figure.figsize':(20,14.60)})
ax = sns.lineplot(x='Date', y='Petroleume_price_in_PKR', data=dframe_pkr, linewidth=5, marker='>', color='g',
                  markersize=10, markevery=4)

# Change line color after May 2022
for index, row in dframe_pkr.iterrows():
    if row['Date'] >= pd.to_datetime('2022-05'):
        ax.lines[0].set_color('#032b5f')

# Set plot title and labels
ax.set(title='Petrol Prices Increase graph of Pakistan from 2020 to 2023',
       xlabel='Date', ylabel='Petrol Price (PKR)')

# Label points on the plot
for x, y in zip(dframe_pkr['Date'],dframe_pkr['Petroleume_price_in_PKR']):
    # Only add label if it has not already been added
    if y not in label_added:
        label_added.append(y)
        # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
        plt.text(x=x, y=y+2, s='{:.00f}'.format(y), color='black')

plt.show()
```

Pakistan from January 2020 to February 2023. # Percentage of Petrol increment as comapre to previous year

```python
import matplotlib.pyplot as plt

# Petrol price data
petrol_prices = [117, 100, 141, 272]

# Calculate percentage increase
percent_increase = [0]
for i in range(1, len(petrol_prices)):
    percent_increase.append(((petrol_prices[i] - petrol_prices[i-1]) / petrol_prices[i-1]) * 100)

# Plot percentage increase over time
years = [2020, 2021, 2022, 2023]
fig, ax1 = plt.subplots()
ax1.bar(years, percent_increase, color='r')
ax1.set_xlabel('Year')
ax1.set_ylabel('Percentage increase',fontsize = 13)
ax1.set_title('Petrol price percentage increase over time',fontsize = 15)

# Add line plot for petrol price
ax2 = ax1.twinx()
ax2.plot(years, petrol_prices, color='black', linestyle='solid', marker = '>', markersize =8)
ax2.set_ylabel('Petrol price (PKR/L)',fontsize = 13)

plt.show()
```

So The bars is representing the percentage increase in petrol prices over the previous year, while the line plot show the actual petrol prices for each year.

# Dollar Exchange Rate into PKR

To analyze the dollar exchange rate into Pakistani rupees over the past 2 years, we used the same dataset that included dollar  exchange rates for Pakistan.

```python
# RUN THIS CODE IN LAST
dframe_pkr['Dollar_rate'] = pd.to_numeric(dframe_pkr['Dollar_rate'], errors='coerce')

# Create a list to store the labels that have already been added
label_added = []

# Plot line graph
sns.set(font_scale=2)
sns.set(rc={'figure.figsize':(20,14.60)})
ax = sns.lineplot(x='Date', y='Dollar_rate', data=dframe_pkr,linewidth=4, marker ='>', color='g',
                  markersize=10,markevery=4)
ax.set(title='Dollar Rate Increase graph of Pakistan from 2020 to 2023'
)
# Label points on the plot
previous_y = None
for x, y in zip(dframe_pkr['Date'],dframe_pkr['Dollar_rate']):
    if previous_y is None or abs(y - previous_y) >= 2:
        # Only add label if it has not already been added
        if y not in label_added:
            label_added.append(y)
            # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
            plt.text(x = x, # x-coordinate position of data label
                     y = y+3, # y-coordinate position of data label, adjusted to be 150 below the data point
                     s = '{:.00f}'.format(y), # data label, formatted to ignore decimals
                     color = '#240a0d') # set colour of line
    previous_y = y
plt.show()
```

"Now we can see that  from January 2020 to February 2023. Our analysis showed that the Dollar exchange rate  ranged from RS.154 to RS.175 between January 2020 and May 2022, But then sharply increased to from May 2022 to February 2023 from RS.180 and now current Price is RS. 268. This significant increase in Dollar prices during the later period is notable and could be further investigated."
